```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
```

```python
class DoublyCircularLinkedList:
    def __init__(self):
        self.head = None

    # Insert at the end of the Doubly Circular Linked List
    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
            new_node.prev = self.head
        else:
            temp = self.head
            while temp.next != self.head:  # Traverse until the last node
                temp = temp.next
            temp.next = new_node
            new_node.prev = temp
            new_node.next = self.head
            self.head.prev = new_node

    # Insert at the beginning of the Doubly Circular Linked List
    def prepend(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
            new_node.prev = self.head
        else:
            temp = self.head
            while temp.next != self.head:  # Traverse until the last node
                temp = temp.next
            temp.next = new_node
            new_node.prev = temp
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node

    # Delete a node from the Doubly Circular Linked List
    def delete(self, key):
        if not self.head:
            print("The list is empty")
            return
        temp = self.head
        # Case when the head is the only node in the list
        if temp.data == key and temp.next == self.head:
            self.head = None
            return
        # Case when the node to delete is the head
        if temp.data == key:
            temp.next.prev = temp.prev
            temp.prev.next = temp.next
            self.head = temp.next
            return
        # Traverse the list to find the node to delete
        while temp.data != key:
            temp = temp.next
            if temp == self.head:
                print("Node not found")
                return
        temp.prev.next = temp.next
        temp.next.prev = temp.prev

    # Print the Doubly Circular Linked List (Forward)
    def print_list(self):
        if not self.head:
            print("The list is empty")
            return
        temp = self.head
        while True:
            print(temp.data, end=" <-> ")
            temp = temp.next
            if temp == self.head:
                break
        print("(head)")

    # Print the Doubly Circular Linked List (Backward)
    def print_reverse(self):
        if not self.head:
            print("The list is empty")
            return
        temp = self.head.prev
        while True:
            print(temp.data, end=" <-> ")
            temp = temp.prev
            if temp == self.head.prev:
                break
        print("(tail)")

    # Search for an element in the Doubly Circular Linked List
    def search(self, key):
        temp = self.head
        if not self.head:
            return False
        while True:
            if temp.data == key:
                return True
            temp = temp.next
            if temp == self.head:
                break
        return False
```

```python
# Create a Doubly Circular Linked List
dcll = DoublyCircularLinkedList()

# Append elements
dcll.append(10)
dcll.append(20)
dcll.append(30)

# Prepend elements
dcll.prepend(5)

# Print list (forward direction)
dcll.print_list()  # Output: 5 <-> 10 <-> 20 <-> 30 <-> (head)

# Print list (reverse direction)
dcll.print_reverse()  # Output: 30 <-> 20 <-> 10 <-> 5 <-> (tail)

# Delete an element
dcll.delete(20)

# Print updated list
dcll.print_list()  # Output: 5 <-> 10 <-> 30 <-> (head)

# Search for an element
print(dcll.search(10))  # Output: True
print(dcll.search(100))  # Output: False
```
